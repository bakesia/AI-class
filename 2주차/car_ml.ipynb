```python
# 라이브러리 로딩
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# -----------------------------
# 1) 데이터 준비(당뇨 데이터)
# -----------------------------

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/기계학습프로그래밍/3주/breast_cancer.csv").dropna()

print(df)

# label만 땜
X = df.drop(columns=["label"])
y = df['label']

print(X)
print(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# -----------------------------
# 2) 모델 구성
# -----------------------------

dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=500)
```

```python
# -----------------------------
# 3) 모델 학습
# -----------------------------

dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)
```

```python
# -----------------------------
# 4) 모델 평가
# -----------------------------

dt_pre = dt.predict(X_test)
rf_pre = rf.predict(X_test)
lr_pre = lr.predict(X_test)

# print(dt_pre)
# print(list(y_test))

dt_acc = accuracy_score(y_test, dt_pre)
rf_acc = accuracy_score(y_test, rf_pre)
lr_acc = accuracy_score(y_test, lr_pre)

print("=== Test Accuracy ===")
print(f"Decision Tree : {dt_acc:.4f}")
print(f"Random Forest : {rf_acc:.4f}")
print(f"Logistic Reg. : {lr_acc:.4f}")
```
